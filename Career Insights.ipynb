```python
# Importing Libraries:

import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the responses data:

df = pd.read_csv(r'C:\Users\asus\Desktop\T1\Responses.csv')
df.head(5)
```

```python
# Renaming the long column names:

df.rename(columns = {'Current job title or future job title':'Careers', 
                     'Career':'Domain', 
                     'Year of Study':'Year',
                     'Which coding language you use mostly in your work.':'Languages', 
                     'College Name':'Colleges',
                    },inplace = True) 
df.head(5)
```

```python
# Number of null cells in different columns:

df.isna().sum()
```

```python
# Stripping the long career names to only first words:

df['Careers'] = df['Careers'].str.split(' ').str[0]
df.groupby('Careers')['ID'].nunique()
```

```python
# Count of different languages:  

df.groupby('Languages')['ID'].nunique()
```

```python
# Count of different colleges:

df.groupby('Colleges')['ID'].nunique()
```

```python
# Striping the long college names to only first words:

df['Colleges'] = df['Colleges'].str.split(' ').str[0]
df.groupby('Colleges')['ID'].nunique()
```

```python
# Deleting the ID and Timestamp column, giving the plotting parameters and checking the datatypes of different columns:

del df['ID']
del df['Timestamp']
plt.rcParams['figure.figsize'] = [15, 5]
df.dtypes
```

```python
# Final Dataset:

df.head(5)
```

# Data Visualization:

```python
# Count of different Career Paths:

plt.figure(figsize = (8,5))
bar1 = sns.countplot(df['Careers'])
bar1.set_xticklabels(bar1.get_xticklabels(), rotation = 90, ha = 'right')
plt.title('Count Careers', size = 24)
plt.xlabel('Careers', size = 18)
plt.ylabel('Count', size = 18)
plt.show()
```

```python
# Count of different Languages Used:

plt.figure(figsize = (8,5))
bar1 = sns.countplot(df['Languages'])
bar1.set_xticklabels(bar1.get_xticklabels(), rotation = 90, ha = 'right')
plt.title('Count Languages', size = 24)
plt.xlabel('Languages', size = 18)
plt.ylabel('Count', size = 18)
plt.show()
```

```python
# Count of different Colleges:

plt.figure(figsize = (8,5))
bar1 = sns.countplot(df['Colleges'])
bar1.set_xticklabels(bar1.get_xticklabels(), rotation = 90, ha = 'right')
plt.title('Count Colleges', size = 24)
plt.xlabel('Colleges', size = 18)
plt.ylabel('Count', size = 18)
plt.show()
```

```python
# Count of different Domains:

plt.figure(figsize = (8,5))
bar1 = sns.countplot(df['Domain'])
bar1.set_xticklabels(bar1.get_xticklabels(), rotation = 90, ha = 'right')
plt.title('Count Domain', size = 24)
plt.xlabel('Domain', size = 18)
plt.ylabel('Count', size = 18)
plt.show()
```

```python
# Bar graph between Career Paths and programing languages used:

plt.style.use('ggplot')

df.groupby(['Careers', 'Languages'])\
      .Careers.count().unstack().plot.bar(legend=True)

plt.show()
```

```python
# Bar graph between Colleges and programing languages used:

plt.style.use('ggplot')

df.groupby(['Colleges', 'Languages'])\
      .Colleges.count().unstack().plot.bar(legend=True)

plt.show()
```

```python
# Bar graph between Career Paths and different Colleges:

plt.style.use('ggplot')

df.groupby(['Careers', 'Colleges'])\
      .Careers.count().unstack().plot.bar(legend=True)

plt.show()
```

```python
# Bar graph between Career Paths and there Domains:

plt.style.use('ggplot')

df.groupby(['Careers', 'Domain'])\
      .Careers.count().unstack().plot.bar(legend=True)

plt.show()
```

```python
#Scatter Plot between Colleges and Career paths:

sns.scatterplot(
    data=df, 
    x="Colleges", y="Careers", hue="Languages")
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
#Scatter Plot between Career paths and programing Languages used:

sns.scatterplot(
    data=df, 
    x="Languages", y="Careers")
plt.rcParams['figure.figsize'] = [15, 10]
```
